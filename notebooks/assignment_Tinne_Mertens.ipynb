```python
# import packages
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# image location
loc_input_img = os.path.join('..', 'data', 'input', 'gekkie.jpg') 

#load image
image = Image.open(loc_input_img)

# convert image to numpy array
np_image = np.array(image)

plt.imshow(np_image)
plt.show()
```

```python
# Flip array to mirror the original
mirrored_image = np.flip(np_image, axis = 1)
plt.imshow(mirrored_image)
plt.show()

# Flip rgb_array so that it is upside down
upside_down_image = np.flip(np_image, axis = (0,1))
plt.imshow(upside_down_image)
plt.show()

# mirror the upside down image
upside_down_mirror_image = np.flip(np_image, axis = 0)
plt.imshow(upside_down_mirror_image)
plt.show()
```

```python
# single array with all image versions
print(np_image.shape,mirrored_image.shape)
joined_image_row1 = np.concatenate((np_image,mirrored_image),axis=1)
joined_image_row2 = np.concatenate((upside_down_mirror_image,upside_down_image),axis=1)
full_image = np.concatenate((joined_image_row1,joined_image_row2),axis=0)
plt.imshow(full_image)
plt.show()
```

```python
# Change all pixels of np_image into blue tints
blue_tint_image = np.copy(np_image)
blue_tint_image[:, :, 0] = 0
blue_tint_image[:, :, 1] = 0

plt.imshow(blue_tint_image)
plt.show()
```

```python
# Change all pixels of np_image into red tints
red_tint_image = np.copy(np_image)
red_tint_image[:, :, 1] = 0
red_tint_image[:, :, 2] = 0

plt.imshow(red_tint_image)
plt.show()
```

```python
# Change all pixels of np_image into green tints
green_tint_image = np.copy(np_image)
green_tint_image[:, :, 0] = 0
green_tint_image[:, :, 2] = 0

plt.imshow(green_tint_image)
plt.show()
```
